```python
# 1. Given a dataset containing eam scores of students,write aprogram using numpy to perform 
# a) cal mean and median of exam scores
# b) compute std deviation and variance

import numpy as np
# Define eam scores
Scores = np.array([75, 76, 80, 83, 86, 98, 92,95])
# a)
mean_score = np.mean(Scores)
median_score = np.median(Scores)

# b)
std_dev = np.std(Scores)
variance = np.var(Scores)

print(f"Mean: {mean_score :.2f}")
print(f"Median: {median_score :.2f}")
print(f"Std_dev: {std_dev :.2f}")
print(f"Variance: {variance :.2f}")
```

```python
# 2) write a prg to display histogram using matplotlib with bins 0,25,50,75,100

import matplotlib.pyplot as plt
import numpy as np

data = np.array([25,30,40,55,60,80,90])
bins = [0, 25, 50, 75, 100]

plt.hist(data, bins=bins, edgecolor = 'violet')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title(' Histogram of Customer Bins')

plt.show()
```

```python
# 4) Write a python prg to display scatter plot for professional exp & salary with sample data

import matplotlib.pyplot as plt
import numpy as np

year_of_exp = np.array([2,3,5,7,8,4,10])
salary = np.array([35,40,65,70,75,38,85])

plt.scatter(year_of_exp, salary, color = 'green', marker ='^')

plt.xlabel('Professional Experience (Years)')
plt.ylabel('Salary (in thousands)')

plt.title(' Experience vs Salary')

plt.show()
```
